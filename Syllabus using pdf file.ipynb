```python
import pdfplumber
import re
from datetime import datetime

def extract_dates_from_syllabus(pdf_file_path):
    # Regular expression to match dates in various formats
    date_patterns = [
        r'\b\d{1,2}/\d{1,2}/\d{2,4}\b',  # Matches MM/DD/YYYY
        r'\b\d{1,2}-\d{1,2}-\d{2,4}\b',  # Matches MM-DD-YYYY
        r'\b\w+ \d{1,2}, \d{4}\b',        # Matches Month DD, YYYY
        r'\b\w+ \d{1,2}\b',               # Matches Month DD
        r'\b\d{1,2} \w+ \d{4}\b'          # Matches DD Month YYYY
    ]

    # Compile all date patterns
    combined_pattern = re.compile('|'.join(date_patterns))
    
    dates_found = []

    # Open the PDF file
    with pdfplumber.open(pdf_file_path) as pdf:
        # Loop through all the pages
        for page in pdf.pages:
            text = page.extract_text()
            if text:  # Check if the page has text
                # Find all matches for date patterns in the text
                matches = combined_pattern.findall(text)
                dates_found.extend(matches)

    # Remove duplicates and sort the dates
    unique_dates = set()
    
    for date in dates_found:
        # Try parsing each date string and add to unique_dates if valid
        for date_format in ['%m/%d/%Y', '%m-%d-%Y', '%B %d, %Y', '%B %d', '%d %B %Y']:
            try:
                parsed_date = datetime.strptime(date, date_format)
                unique_dates.add(parsed_date)  # Store the parsed date
                break  # Stop checking once a valid format is found
            except ValueError:
                continue  # Move to the next date format if it fails

    # Sort dates and convert back to string format for output
    sorted_dates = sorted(unique_dates)
    formatted_dates = [date.strftime('%B %d, %Y') for date in sorted_dates]

    return formatted_dates

# Specify the path to the PDF syllabus file
pdf_file_path = "EGN3443 Fall 2024 Li & Gobes-Ryan.pdf"
# Extract and print key academic dates
academic_dates = extract_dates_from_syllabus(pdf_file_path)

# Display the results with proper formatting
print("Key Academic Dates:")
if academic_dates:
    for date in academic_dates:
        print(f"- {date}")
else:
    print("No academic dates found.")
```
